```python
import pandas as pd
```

## 월별 hs코드별 수출합계

```python
def export_sum(start_date, end_date):
    df_month=pd.read_excel('data_month.xlsx',)
    filtered_dataframe = df_month[(df_month['기간'] >= start_date) & (df_month['기간'] <= end_date)]
    월별HS코드별합계=filtered_dataframe.groupby(['기간','품목'])[['수출금액','수입금액','무역수지']].sum()
    월별HS코드별합계=월별HS코드별합계.sort_values('기간',ascending=False).reset_index()

    월별HS코드별합계=월별HS코드별합계.fillna(method='ffill')#nan넣기
    월별HS코드별수출합계=월별HS코드별합계.pivot(index='품목',columns='기간',values='수출금액')
    월별HS코드별수출합계=월별HS코드별수출합계[sorted(list(월별HS코드별수출합계.columns),reverse=True)].reset_index() #순서변경

    index=[5,6,7,12,13,14,15,16,17,0,8,9,2,1,3,4,10,11]
    index_name=['산업용 로보트','산업용 로보트','기타 산업용 로봇','과용접 로봇','봉합용접 로봇','바트용접 로봇','저항용접 로봇','교류아크용접 로봇','기타 아크용접 로봇','도장 로봇','전자부품장차기(SMT)','기타 전자부품장착기(SMT)','적재용 로봇','무인운반로봇(AVG)','무인운반로봇(8427제외)','기타 무인운반로봇','로봇청소기','기타로봇청소기']
    월별HS코드별수출합계=월별HS코드별수출합계.reindex(index)
    월별HS코드별수출합계.insert(1,'구분',index_name)
    월별HS코드별수출합계=월별HS코드별수출합계.fillna(0)
    return 월별HS코드별수출합계

월별HS코드별수출= export_sum(
start_date = float(input("시작날짜 (ex. 2023.01)")),
end_date = float(input("종료날짜 (ex. 2023.01)"))
)
월별HS코드별수출.to_excel('월별HS코드별수출합계.xlsx',index=False) #엑셀로 저장
월별HS코드별수출
  #  월별HS코드별수출합계.to_excel('월별HS코드별수출합계.xlsx',index=False)
   # 월별HS코드별수출합계
```

## 연도별 국가별 수출입 합계

```python
def year_contry_sum(contry, start_year, end_year):
    df_year=pd.read_excel('data_year.xlsx',)
    filtered_dataframe = df_year[df_year['국가'].isin(contry)&(df_year['기간'] >= start_year) & (df_year['기간'] <= end_year)]
    국가별합계=filtered_dataframe.groupby(['기간','국가'])[['수출금액','수입금액','무역수지']].sum().sort_values(['기간','수출금액'],ascending=False)
    return 국가별합계



국가별 = year_contry_sum(
contry = list(input("조회하고 싶은 나라를 입력하세요 (띄어쓰기로 구분)").split()),
start_year = int(input("시작 년도")),
end_year = int(input("종료 년도"))
)

국가별.to_excel('연도별국가별수출입합계.xlsx',index=False) #엑셀로 저장
국가별
# 국가별합계.to_excel('국가별합계.xlsx')
```

## 연도별 HS코드별 총 수출입 무역수지 통계

```python
def year_sum(start_year, end_year):
    df_year=pd.read_excel('data_year.xlsx',)
    filtered_dataframe = df_year[(df_year['기간'] >= start_year) & (df_year['기간'] <= end_year)]
    연도별HS코드별합계=filtered_dataframe.groupby(['기간','품목'])[['수출금액','수입금액','무역수지']].sum()
    연도별HS코드별합계=연도별HS코드별합계.sort_values('기간',ascending=False)
    return 연도별HS코드별합계

연도별HS코드별 = year_sum(
    start_year = int(input("시작년도")),
    end_year = int(input("종료년도"))
)

연도별HS코드별.to_excel('연도별HS코드별합계.xlsx',index=False) #엑셀로 저장
연도별HS코드별
```
